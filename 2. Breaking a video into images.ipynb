#### We will break down video into frames using cv2.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PATH = '/Users/gursharan/Desktop/sign_lang.mp4'
```

```python
cap = cv2.VideoCapture(PATH)
image_frames = [];
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    image_frames.append( frame );

cap.release()
cv2.destroyAllWindows()
```

```python
print( len(image_frames ) )
```

```python
_ = plt.imshow( image_frames[1000] )
```
